```python
import pandas as pd
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS

def analyze_and_generate_wordcloud(csv_file, stopwords=None):
    # Read CSV file
    df = pd.read_csv(csv_file)

    # Combine all text data into a single string (using data from the second column)
    all_text = ' '.join(df.iloc[:, 1].dropna())

    # Tokenize the text into words
    words = all_text.split()

    # Remove stopwords
    if stopwords is not None:
        words = [word for word in words if word.lower() not in stopwords]

    # Count word frequencies
    word_freq = Counter(words)

    # Create a DataFrame from word frequencies
    word_freq_df = pd.DataFrame(list(word_freq.items()), columns=['Word', 'Frequency'])

    # Sort DataFrame by frequency in descending order
    word_freq_df = word_freq_df.sort_values(by='Frequency', ascending=False)

    # Display word frequencies as text
    print("Word Frequency Table:")
    print(word_freq_df)

    # Save word frequencies to CSV file
    word_freq_df.to_csv('word_frequencies.csv', index=False)

    # Generate a word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

    # Plot the word cloud
    plt.figure(figsize=(8, 4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud')
    plt.show()

# Set the path to your CSV file
csv_file_path = 'output_results1.csv'

# Set custom stopwords if needed
# Set custom stopwords if needed
custom_stopwords = set([
    "your", "custom", "stopwords", "man", "video", "people",
    "man's", "camera", "takes", "seems", "seen", "woman","front","starts","appears","again,", "it.","group","piece","continues","festures","see"
    "standing", ",", "around", "", "Overall",
])

# Call the function to analyze and generate word cloud
analyze_and_generate_wordcloud(csv_file_path, stopwords=STOPWORDS.union(custom_stopwords))


# Call the function to analyze and generate word cloud
analyze_and_generate_wordcloud(csv_file_path, stopwords=STOPWORDS.union(custom_stopwords))
```
